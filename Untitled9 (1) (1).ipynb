```python
import pandas as pd

data = {
    'Department':['HR','IT','HR',None],
    'Salary':[30000,50000,30000,40000]
}

df = pd.DataFrame(data)

df.drop_duplicates(inplace=True)
df.dropna(inplace=True)

print(df)
```

```python
import pandas as pd

data = {
    'Department':['HR','IT','Sales','HR'],
    'Attrition':['Yes','No','Yes','No']
}

df = pd.DataFrame(data)

print(df.groupby('Department')['Attrition'].value_counts())
```

```python
import pandas as pd

data = {
    'Department':['HR','IT','Sales'],
    'Salary':[30000,50000,25000]
}

df = pd.DataFrame(data)

print(df.groupby('Department')['Salary'].mean())
```

```python
import pandas as pd

data = {
    'Department':['HR','IT','Sales'],
    'Experience':[2,5,1]
}

df = pd.DataFrame(data)

print(df.groupby('Department')['Experience'].mean())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data = {
    'Department':['HR','IT','Sales','HR'],
    'Attrition':['Yes','No','Yes','No']
}

df = pd.DataFrame(data)

df['Department'].value_counts().plot(kind='bar')
plt.title("Employees by Department")
plt.show()

df['Attrition'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title("Attrition Distribution")
plt.ylabel("")
plt.show()
```
